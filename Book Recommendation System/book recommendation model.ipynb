Recommendation based on Popularity

```python
ratings_with_name= ratings.merge(books, on='ISBN')
ratings_num_df= ratings_with_name.groupby('Book-Title'). count()['Book-Rating'].reset_index()
ratings_num_df.rename(columns={'Book-Rating': 'ratings_num'}, inplace= True)
ratings_num_df
                    
```

```python
avg_ratings_df= ratings_with_name.groupby('Book-Title'). count()['Book-Rating'].reset_index()
avg_ratings_df.rename(columns={'Book-Rating': 'avg_ratings'}, inplace= True)
avg_ratings_df
```

```python
popular_df= ratings_num_df.merge(avg_ratings_df, on='Book-Title')
popular_df
```

```python
popular_df= popular_df[popular_df['ratings_num'] >= 250].sort_values('avg_ratings', ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','ratings_num','avg_ratings']]
```

```python
popular_df['Image-URL-M'][0]
```

Colaborative Filtering 

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
literate_users = x[x].index

filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(literate_users)]
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    #fetch index
    index= np.where(pt.index==book_name)[0][0]
    similar_items= sorted(list(enumerate(similarity_scores[index])), key=lambda x:x[1],reverse=True)[1:5]
    data= []
    for i in similar_items:
        item=[]
        temp_df= books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)
    return data
```

```python
recommend('1984')
```
